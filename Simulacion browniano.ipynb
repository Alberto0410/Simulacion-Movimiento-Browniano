```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

1. Dado $N = 100$, genere una muestra aleatoria $\{X_i\}_{i = 1}^{N}$ de variables aleatorias con distribución
$$P(X_i = 1) = \frac{1}{2} = 1 - P(X_i = -1)$$
Obtenga a partir de allí una trayectoria ($S_k, k \in \{0, 1, \ldots, N\}$) y grafíquela

```python
np.random.seed(24)
random.seed(24)

def symmetric_random_walk(N, start = 0, plot = False):
    '''Función que simula la caminata aleatoria simétrica de N pasos.
    Entrada:
        N: número de pasos
        start: valor inicial
        plot: si es True, grafica la caminata, en caso contrario, devuelve el vector de la caminata
    Salida:
        random_walk: vector de la caminata aleatoria'''
    
    #creamos las observaciones de las X_i
    observaciones = [random.choice([-1, 1]) for _ in range(N)]
    random_walk = np.array(observaciones)
    random_walk = np.cumsum(random_walk)

    #agregamos el inicio y creamos el vector del tiempo
    random_walk = np.concatenate(([start], random_walk))
    tiempo = range(N + 1)
    
    #graficamos
    if plot:
        plt.figure(figsize = (15, 8))
        plt.plot(tiempo, random_walk, linewidth = 2, color = '#31A74A')
        plt.xlabel('Tiempo')
        plt.ylabel('Posición')
        plt.title(f'Caminata aleatoria simétrica con N = {N}')
        plt.grid()
        plt.show()

    else:
        return random_walk
```

```python
symmetric_random_walk(100, plot = True)
```

2. Para $r = 10; 20; 50$, grafique $\left(W_{t/r}^{(r)}\right), t \in [0, N]$ a partir de $\left(W_{k/r}^{(r)}, k \in \{0, 1, \ldots, N\}\right)$ usando interpolación lineal. 

```python
np.random.seed(24)
random.seed(24)

def scaled_simetric_rw(N, r, plot = False):
    '''Función que simula la caminata aleatoria simétrica reescalada en N pasos.
    Entrada:
        N: número de pasos
        r: escala
        plot: si es True, grafica la caminata, en caso contrario, devuelve el vector de la caminata
    Salida:
        random_walk: vector de la caminata aleatoria'''
    
    #primero obtenemos una caminata aleatoria de N pasos
    random_walk = symmetric_random_walk(N)

    #la reescalamos
    random_walk_scaled = random_walk / np.sqrt(r)
    tiempo = range(N + 1)

    #graficamos
    if plot:
        plt.figure(figsize = (15, 8))
        plt.plot(tiempo, random_walk_scaled, linewidth = 2, color = '#31A74A')
        plt.xlabel('Tiempo')
        plt.ylabel(r'$W_{k/r}^{(r)}$')
        plt.title(f'Caminata aleatoria simétrica reescalada con N = {N} y r = {r}')
        plt.grid()
        plt.show()
    
    else:
        return random_walk_scaled
```

```python
r_list = [10, 20, 50]

for r in r_list:
  scaled_simetric_rw(100, r, plot = True)
```

3. Finalmente, grafique $\left(W_{t}^{(N)}, t \in [0, 1]\right)$ usando el mismo método que en el ítem anterior.

```python
def scaled_simetric_rw01(N, r, plot = False):
    '''Función que simula la caminata aleatoria simétrica reescalada en N pasos
    en el intervalo [0, 1].'''

    #obtenemos la caminata aleatoria reescalada
    random_walk_scaled = scaled_simetric_rw(N, r, plot = False)

    #reescalamos el tiempo 
    tiempo = np.linspace(0, 1, N + 1)

    #graficamos
    if plot:
        plt.figure(figsize = (15, 8))
        plt.plot(tiempo, random_walk_scaled, linewidth = 2, color = '#31A74A')
        plt.xlabel('Tiempo')
        plt.ylabel(r'$W_{k/r}^{(r)}$')
        plt.title(f'Caminata aleatoria simétrica reescalada con N = {N} y r = {r}')
        plt.grid()
        plt.show()
```

```python
r_list = [10, 20, 50]

for r in r_list:
  scaled_simetric_rw01(100, r, plot = True)
```
